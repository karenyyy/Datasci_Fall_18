# Linear Regression


```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# optional, to look nice
from matplotlib import style; style.use('ggplot')
plt.rcParams['figure.figsize'] = [8, 6]
```

```python
# load data
data = pd.read_csv('linear_regression_data.csv')

# set up variables
X = data.X
y = data.y
m = y.size
```

```python
# visualize data
plt.scatter(X, y, color='blue')
plt.show()
```

```python
# add column of ones to X to account for y-intercept
X = np.hstack((np.ones((m, 1)), 
               X.values.reshape(m, 1)))

# make y 2-Dimensional to match X
y = y.values.reshape(m, 1)

# initialize parameters slope and y-intercept to 0
theta = np.zeros((2, 1))
```

```python
# make predictions for all points X
def predict():
    return np.matmul(X, theta)
```

```python
# compute cost averaged over all points
def cost():
    return (1 / (2 * m)) * np.sum((predict() - y) ** 2)
```

```python
# compute slope (derivative of cost with respect to theta)
def gradient():
    return (1 / m) * np.matmul(X.T, (predict() - y))
```

```python
# set gradient descent parameters
num_iterations = 1500
learning_rate = 0.01
```

```python
# perform gradient descent
for i in range(num_iterations):
    theta -= learning_rate * gradient()
    plt.scatter(i, cost(), color='red')

# show how cost decreases with each adjustment
plt.show()
```

```python
# create endpoints for plotting
X_pt = [min(X[:, 1]), max(X[:, 1])]
y_pt = [min(predict()), max(predict())]

# plot line of best fit, learned through gradient descent
plt.scatter(X[:, 1], y, color='blue')
plt.plot(X_pt, y_pt, color='red')
plt.show()
```
