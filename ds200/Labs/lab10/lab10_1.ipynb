```python
import pandas as pd
import numpy as np
```

```python
to_numeric = {
                    'Positive':1,
                    'Negative':-1,
                    'Neutral': 0,
                    'not-relevant':-2
                    }
```

```python
def lab10(climate, newlabel, supportive):
    for key, val in to_numeric.items():
        for label in climate.columns[2:]:
            climate[label] = climate[label].replace(key, val)
    climate[newlabel] = climate[climate.columns[2:]].apply(lambda x: np.sum(x.values), axis=1)
    climate[supportive] = climate.iloc[:, -1].apply(lambda x: 1 if x>=1 else 0)
    return climate
```

```python
climate = pd.read_csv('Climate1Tags.csv', sep=',')
lab10(climate, 'newlabel', 'supportive')
```
